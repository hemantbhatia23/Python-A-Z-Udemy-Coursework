```python
 #Building Dashboards
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = 8,5
```

```python
sns.set_style('dark',{'axes.facecolor':'black'})
f,axes = plt.subplots(2,2,figsize=(15,15))
#Plot 0,0
k1 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,\
                 ax=axes[0][0],shade=True,shade_lowest=True,\
                 cmap='inferno')
k1o = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,\
                 ax=axes[0][0],cmap='cool')
#Plot 0,1
k2 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,\
                 ax=axes[0][1],shade=True,shade_lowest=True,\
                cmap='inferno')
k2o = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,\
                 ax=axes[0][1],
                cmap='cool')

#Plot 1,0
z = sns.violinplot(data=movies,x='Year',y='BudgetMillions',\
                   ax=axes[1][0],palette='YlOrRd')
#Plot 1,1
k1 = sns.kdeplot(movies.CriticRating,movies.AudienceRating,\
                 shade=True,shade_lowest=False,cmap='Blues_r',\
                 ax=axes[1][1])
kld = sns.kdeplot(movies.CriticRating,movies.AudienceRating,\
                 cmap='gist_gray_r',ax=axes[1][1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
BudgetMil =[]
myLabels =[]
for g in movies.Genre.cat.categories:
    BudgetMil.append(movies[movies.Genre == g].BudgetMillions)
    myLabels.append(g)

sns.set_style('whitegrid')
fig,ax = plt.subplots()
fig.set_size_inches(11.7,8.27)
plt.hist(BudgetMil,bins=30,stacked=True,rwidth=1,label =myLabels,color=['red','Salmon','Blue','Green','Magenta','Purple','Black'] )
plt.title('Movie Budget Distribution',fontsize=25,color='DarkBlue'\
         ,fontname='arial')
plt.ylabel('Number Of Movies',fontsize=25,color='red')
plt.xlabel('Budget',fontsize=25,color='green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={'size':20},frameon=True,fancybox=True\
          ,framealpha=1,shadow=True,)
plt.show()
```
